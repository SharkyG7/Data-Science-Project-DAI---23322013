```python
import ast

def convert_list(entries):
    temp = []
    entries = ast.literal_eval(entries)
    for x in entries:
        temp.append(x["name"])

    return temp
    
df["genres"] = df["genres"].apply(convert_list)
df["spoken_languages"] = df["spoken_languages"].apply(convert_list)
df["production_companies"] = df["production_companies"].apply(convert_list)
df["production_countries"] = df["production_countries"].apply(convert_list)
```

3.6 | Handeling Unrealistic Values

```python
# Some budget and revenue entries are zero which is unrealistic.
zero_budget = df.loc[df["budget"] == 0]
zero_budget.head()
```

3.5 | Outliers

```python
def detect_outliers(df: pd.DataFrame, column: str):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return f"{column} : {outliers[column].size}"

columns_to_check = ['popularity', 'budget', 'revenue', 'runtime',  'vote_count', 'vote_average',
        'budget_adj', 'revenue_adj']

# Detect outliers for each column and store the results in a list
all_outliers = [detect_outliers(df,column) for column in columns_to_check]
all_outliers
```

### Insights

1.   Outliers are not very significant in the data.
2.   Therefore no need to remove anything.

#### Ques11. What are the top 10 movies by profit?

```python
def get_top_10_movies(df, by_col):

    top_movies = df.sort_values(by=by_col, ascending=False)[:10]
    top_movies["title_with_year"] = top_movies["original_title"] + " (" + top_movies["release_year"].astype(str) + ")"

    plt.figure(figsize=(10, 6))
    p = sns.barplot(x=top_movies[by_col], y=top_movies["title_with_year"], palette="viridis")

    plt.xlabel(f"{by_col.title()}")
    plt.ylabel("Movie Title")
    plt.title(f"Top 10 Movies by {by_col.title()}")

    for container in p.containers:
        p.bar_label(container, color="black", bbox = {'boxstyle': 'square', 'facecolor': 'cyan', 'edgecolor': 'red'})

    plt.show()
```

##### Imdb Weighted Rating Formula

```python
# We Already Have V and R
# C is too easy
C = df["vote_average"].mean()
print("Mean votes :", C)

# M is the minimum votes

M = df["vote_count"].quantile(.9)
print("Minimum Votes :", M)
# So i will take only the top 10 percent of the dataset to plot it / above than (582)

#Filter the dataset acordding to the m

top_10p_movies = df.loc[df["vote_count"] >= M]

top_10p_movies
```

#### Ques15. What are the top 10 movies by score?

```python
def weighted_rating(df, M=M, C=C):
    v = df['vote_count']
    R = df['vote_average']

    # Calculation based on the IMDB formula
    return (v/(v+M) * R) + (M/(M+v) * C)

# Applying the function to DataFrame and we chosse the axis=1 because we need it to apply on the rows
top_10p_movies["score"] = top_10p_movies.apply(weighted_rating,axis=1)


get_top_10_movies(top_10p_movies,"score")
```

*  Most of the movies are old.
*  The Dark Knight , Inception, Interstellar and The Lord of the Rings are the newest movies.

#### Ques22. Is there a correlation between revenue and budget?

```python
def plot_and_correlate(df , x_col, y_col):

    correlation = df[x_col].corr(df[y_col])
    print(f"Pearson correlation coefficient between {x_col} and {y_col}: {correlation:.2f}")

    fig, ax = plt.subplots(figsize=(8, 6))
    sns.regplot(x=pd.to_numeric(df[x_col]), y=pd.to_numeric(df[y_col]), line_kws={"color": "r"}, ax=ax)
    ax.set_xlabel(x_col.capitalize())
    ax.set_ylabel(y_col.capitalize())
    ax.set_title(f"Relationship between {x_col.capitalize()} and {y_col.capitalize()}")

    plt.show()

plot_and_correlate(df=df,x_col="revenue_adj",y_col="budget_adj")
```

##### Insights
*  There is a relation between budget and revenue.
*  But because there is a lot of zeros the graph is not the best.
*  As the budget increase the revenue also increase
